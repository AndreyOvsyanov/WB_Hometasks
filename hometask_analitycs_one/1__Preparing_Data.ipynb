## 1

Пример плотности вероятности для равномерного распределения на интервале [a,b].
Функция плотности вероятности для этого распределения равна: f(x) = 1 / (b - a)
Теперь давайте напишем код на Python для генерации массива чисел из этого распределения и построения графика плотности вероятности и гистограммы:

```python
import numpy as np
import matplotlib.pyplot as plt

# Функция для плотности вероятности равномерного распределения
def uniform_pdf(x, a, b):
    return 1 / (b - a)

# Генерация массива чисел из равномерного распределения на интервале [a, b]
a, b = 2, 8
data = np.random.uniform(a, b, 1000)

# Построение гистограммы
plt.hist(data, bins=30, density=True, alpha=0.5, color='b', label='Гистограмма')

# Генерация значений для построения графика плотности вероятности
x = np.linspace(a, b, 100)
y = [uniform_pdf(x, a, b)]*len(x)

# Построение графика плотности вероятности
plt.plot(x, y, 'r-', label='Плотность вероятности')

# Добавление подписей и легенды
plt.title('Равномерное распределение на интервале [2, 8]')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.legend()

# Отображение графика
plt.show()
```

## 2



Пример дискретной случайной величины - количество выпадений определенного значения на двух игральных костях. Значениями, которые может принимать эта СВ, будут числа от 2 до 12, а закон распределения может выглядеть следующим образом:

| Значение СВ | Вероятность |
|----------|----------|
| 2     | 1/36   |
| 3     | 2/36   |
| 4     | 3/36   |
| 5     | 4/36   |
| 6     | 5/36   |
| 7     | 6/36   |
| 8     | 5/36   |
| 9     | 4/36   |
| 10    | 3/36   |
| 11    | 2/36   |
| 12    | 1/36   |



```python
import numpy as np

# Значения и вероятности для суммы значений на двух игральных костях
values = np.arange(2, 13)
probabilities = np.array([1, 2, 3, 4, 5, 6, 5, 4, 3, 2, 1]) / 36

# Генерация массива чисел, симулирующих наблюдение дискретной случайной величины
random_sums = np.random.choice(values, size=1000, p=probabilities)

# Вывод результатов
print("Сгенерированные суммы:")
print(random_sums)

# Визуализация гистограммы
import matplotlib.pyplot as plt

plt.hist(random_sums, bins=np.arange(1.5, 13.5, 1), density=True, alpha=0.5, color='b', edgecolor='black')
plt.title('Распределение сумм значений на двух игральных костях')
plt.xlabel('Значение')
plt.ylabel('Вероятность')
plt.show()
```
